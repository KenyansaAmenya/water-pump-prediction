## STEP 3: Hypothesis Testing and Correlation Analysis  
**Hypothesis testing** is a statistical method used to determine whether there's enough evidence in data to suport a claim about a r/ship between variables.   

Hypothesis testing helps answer the questions such as:  
- Does a feature influence the target?
- Is this feature useful for prediction?
- Are these differences occuring by chance?  

For supervised learning,especially classification, hypothesis tests help in:  
- Feature selection  
- EDA(Check if the data makes statistical sense)
- Understanding variable r/ships  
- Reducing irrelevant features  

##### The core idea of hypothesis testing  
Hypothesis testing start with 2 competing statements:  
_Null Hypothesis(H0)_ > This is the default assumption that there's no effect, no relationship, no difference btn groups.   
exampe: Gender and Loan Approval are independent(means that that the feature doesn't affect the target(loan approval) )  
_Alternative Hypothesis(H1)_ > This is what we are trying to prove.   
example: Gender and Loan Approval are associated(Gender might contain predictive information on the Loan Approval)  

_Why hypothesis testing?_ They help us to get statistical evidence instead of guessing.  

__Definition of terms__  
- alpha(level of significance) - Alpha defines the cutoff for 'statistical significance'.(Think of it as the chance given to prove yourself) 0.05(5%), 0.10/10%, 0.01/1% 
- p-value - A p-value is the probability of observing results at least as extreme as the current data assuming the null hypothesis is true.  
- Type I Error(False Positive) - When we reject a null hypothesis that is true  
- Type II error(False Negative) - When we fail to reject a null hypothesis.  
- Confidence interval - It provides a range of plausible values for a population parameter, typically estimated at 95%/99% confidence level.    

|Condition|Decision|
|---------|--------|
|p < alpha|Reject H0|
|p >=alpha| Fail to reject H0|  

_Why we don't accept alternative hypothesis_ The core logic of science and stats is that we can never prove a claim is 100% true, we can only prove a claim is false.  

**_How do you choose the best hypothesis test_**  
|Feature type|target variable|recommended test|
|------------|--------------|-----------------|
|Numerical|Numerical|correlation/Regression|
|Numerical|Categorical|t-test/Anova/MannUWhitney/KruskalWallis|  
|Categorical|Categorical|Chi-Square test|

When the feature is numerical and the target is categorical the goal is usually to determine:  
"Do the numerical values differ significantly across the target classes?"  

_Examples_  
|Numerical|Categorical target| 
|---------|------------------|
|Income|Loan Approval|
|Age|Disease Status|
|Exam Scores|Pass/Fail|  

- The core statistical question is we want to determine whether the numerical feature behaves differently across categories.  
_Example_  
_Do approved and rejected applicants have different average incomes?_  
Null Hypothesis(H0) > Approved and rejected applicants income have no significant difference.  
Alternative Hypothesis(H1) >  Approved and rejected applicants incomes have a significant difference.  

The choice of test to be used depends mainly on:  
- Number of target categories
    - Binary classification(t-test/z-test)
    - Multiclass(ANOVA)
- Sample size  
    -If sample size is less than 30 (n < 30) use t-test, if sample size is geater than 30(n >= 30)
    use z-test  
- Whether the population variance is known  
    - Use a z-test if the population variance is know else use t-test irregardless of the sample size. 

- Distribution of the dataset  
 - Normally distributed data use t-test/z-test/ANOVA
 - In a case of data not normally distributed use 
    - Binary Mann-Whitney U test(replaces t-test / z-test)
    - Multiclass Kruskal-Wallis test(ANOVA)  

#### Binary Distribution  
Supposed our target:  
|Loan Status|
|-----------|
|Approved|
|Denied|  

and the featue income, we compare:  
_mean income of approved applicants_ vs _mean income of rejected applicants_  

__Hypothesis__  
Null(H0): mean1(approved) = mean2(denied) There's no significant difference between the means    
Alternative(H1): mean1(approved) != mean2(denied)  There's a significant fdifference between the means  

__The t-test__   
The t-test determines whether two sample means differ significantly. 
It is the most used test when conducting test on a numerical vs binary categorical target because in real world, population variance is unknown.  

__Types of t-test__  
|Type         |Usage        |
|--------------|------------|
|One-sample t-test|Compare sample mean to a fixed value|
|Independent two-sample t-test|Compare two independent groups|
|Paired t-test|Compare same subjects before/after|   

For ML classification Independent two-sample ttest is the most common.  
Formula:   

$$t = \frac{\bar{x}_1 - \bar{x}_2}{SE}$$
 


Where:  
bar{x}_1 = mean of group 1  
bar{x}_2 = mean of group 2  
SE = standard error  

__Interpretation__  
_Large t-value_:  
Means:  
group means differ substantially thus Reject null hypothesis.  

_Small t-value_  
Means:  
group means are similar thus fail to reject null hypothesis. 

__Z-test__  
A z-test is similar to a t-test but assumes:  
- Pupulation variance is known and sample size is very large(n>=30)  

_When to use z-test_  
- When you have a numerical feature  
- Binary categorical feature  
- Population variance is known or very large sample size(n>=30)  


#### Multiclass Classification  
Suppose target:  
|Target|

|non-functional|
|functional need repair|   

Now we compare more than 2 groups means, so we use ANOVA(Analysis Of Variance).   
__Hypothesis__  
Null hypothesis(H0): mean1=mean2=mean2 There's no significant difference in the group means.  
Alternative Hypothesis(H1):  Atleast one mean differs.

#### Categorical vs Categorical target  
Chi-Sqaure test of independence  
_Why Chi Square?_ The Chi-Square test is specifically designed to determine whether two categoical variables are associated.  
It compares:  
What we actually observed in the data vs What we would expect if there's no r/ship.  

_Intuition behind Chi-Square_  
Suppose we have:   
|Education Level|Approved|Rejected|
|---------------|--------|--------|
|High School|      20|       80|
|Degree|70       |       30|  

You immediately notice:  
Degree holders are approved more often as compared to highschool applicants.  
but:  
Is this a real relationship? Or a random chance/coincidence?

```python
# Looking relationship between the categorical features and status_group(target)
cat_cols = cleaned_df.select_dtypes(include=['object', 'bool']).columns.tolist()
cat_cols

alpha = 0.05 
for col in cat_cols:
    
    null = f"{col} and status_group(target) are independent"
    alternative = f"{col} and status group(target) are not independent"  

    contingency_table = pd.crosstab(cleaned_df[col], cleaned_df["status_group"])
    chi2, p, dof, expected = chi2_contingency(contingency_table)
    print("="*80)
    print(f"The p-value for '{col}' is {p:.55f}") 
    if p < alpha:
        print(f"Reject the null hypothesis that {null}")
    else:
        print(f"Fail to reject the null hypothesis {null}")
```

### Numerical Features(Numerical vs categorical)
#### What if the Data if NOT normally distributed
- Many real-world ml datasets violate the normality assumption
- When:
     - numerical features are not normally distributed,
     - a numerical feature contains strong outliers,
     - or sample sizes are small
     the standard parametric tests such as, t-test, z-test, ANOVA may become unreliable.
in such cases we use:
**Non-parametric Tests**
These tests do not assume normality, and are more robust to skewed distributions and outliers.

Parametric Vs Non-Parametric Tests
|Type|Assumes Normality?|Use Means?|
|----|------------------|----------|
|Parametric()t-test/z-test/ANOVA|Yes|Yes|
|Non-parametric|No|Uses ranks|


##### If Data is NOT Normal
The replacements depends on:
**Number classes**
1. Binary categorical target > Mann-Whitney U Test/Wilcoxon Rank-Sum Test
- Use when you have numerical feature and two target classes(binary) e.g Spam or not spam
- Non-normal data
_Why Mann-Whitney works_ Instead of comparing means it compares ranks i.e values are ordered from
 smallest to largest, the test evaluates whether one group tends to have higher values.

NULL(H0): The distributions are equal.
Alternative(H1): the distribution differ. 

2. Multiclass target > Kruskal-Wallis Test(Replacement for ANOVA)
Hypothesis:
_Null(H0)_: All groups come from the same distribution.
_Alternative(H1)_: Atleast one group differs.

**Summary table**
|Scenario|Normal Data|Non-Normal Data|
|--------|-----------|---------------|
|Numerical vs 2-class categorical(Binary)|t-test/z-test|Mann-Whitney U|
|Numerical vs multiclass categorical|ANOVA|Kruskal Wallis|

```python
# Hypothesis testing for numerical features vs categorical target using Kruskal-Wallis H-test  
num_cols = cleaned_df.select_dtypes(include=[np.number]).columns.tolist()

for col in num_cols:
    alpha = 0.05  
    null = f"The distribution of the numerical feature '{col}' is the same across all status groups/do not differ"
    alternative = f"The distribution of the numerical feature '{col}' differs across at least one status group"  

    groups = [group[col].values for name, group in cleaned_df.groupby("status_group")]
    stat, p = kruskal(*groups)
    print("="*80)
    print(f"The p-value for '{col}' is {p:.55f}")
    if p < alpha:
        print(f"Reject the null hypothesis that {null}") 
    elif p >= alpha: 
        print(f"Fail to reject the null hypothesis that {null}")
```
